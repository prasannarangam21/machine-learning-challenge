```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
# Remove Space for `FALSE POSITIVE` category
mask = df["koi_disposition"] == "FALSE POSITIVE"
df.loc[mask, "koi_disposition"] = "False_Positive"
df["koi_disposition"]
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_fpflag_co', 'koi_fpflag_nt', 'koi_fpflag_ss', 'koi_model_snr', 'koi_prad','koi_fpflag_ec']]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df["koi_disposition"]
X = df.drop(columns=["koi_disposition"])
X
```

```python
# from sklearn.model_selection import train_test_split
# X = df.drop(columns=['koi_disposition'])
# y = df['koi_disposition']
# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Model



```python
# from sklearn.linear_model import LogisticRegression
# classifier = LogisticRegression(solver='liblinear',multi_class='auto')
# classifier.fit(X_train_scaled,y_train)
```

```python
# Create the SVC Model
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model
```

```python
# Train the Model
model.fit(X_train_scaled,y_train)
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
# from sklearn.model_selection import GridSearchCV
# param_grid = {'C': [1, 5, 10],
#               'penalty': ['l1', 'l2']}
# grid = GridSearchCV(classifier, param_grid, cv=3, verbose=3)
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 10, 100, 1000],
              'gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
predictions = grid.predict(X_test_scaled)
```
